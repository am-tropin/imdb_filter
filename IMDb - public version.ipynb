```python
# title.ratings.tsv.gz – Contains the IMDb rating and votes information for titles

# tconst (string) - alphanumeric unique identifier of the title
# averageRating – weighted average of all the individual user ratings
# numVotes - number of votes the title has received

title_ratings = pd.read_csv('title.ratings.tsv',sep='\t')
title_ratings.head(5)
```

```python
# name.basics.tsv.gz – Contains the following information for names:

# nconst (string) - alphanumeric unique identifier of the name/person
# primaryName (string)– name by which the person is most often credited
# birthYear – in YYYY format
# deathYear – in YYYY format if applicable, else '\N'
# primaryProfession (array of strings)– the top-3 professions of the person
# knownForTitles (array of tconsts) – titles the person is known for

name_basics = pd.read_csv('name.basics.tsv',sep='\t')
name_basics.head(5)
```

```python
# title.crew.tsv.gz – Contains the director and writer information for all the titles in IMDb. Fields include:

# tconst (string) - alphanumeric unique identifier of the title
# directors (array of nconsts) - director(s) of the given title
# writers (array of nconsts) – writer(s) of the given title

title_crew = pd.read_csv('title.crew.tsv',sep='\t')
title_crew.head(5)
```

```python
# title.episode.tsv.gz – Contains the tv episode information. Fields include:

# tconst (string) - alphanumeric identifier of episode
# parentTconst (string) - alphanumeric identifier of the parent TV Series
# seasonNumber (integer) – season number the episode belongs to
# episodeNumber (integer) – episode number of the tconst in the TV series

title_episode = pd.read_csv('title.episode.tsv',sep='\t')
title_episode.head(5)
```

```python
# title.akas.tsv.gz - Contains the following information for titles:

# titleId (string) - a tconst, an alphanumeric unique identifier of the title
# ordering (integer) – a number to uniquely identify rows for a given titleId
# title (string) – the localized title
# region (string) - the region for this version of the title
# language (string) - the language of the title
# types (array) - Enumerated set of attributes for this alternative title. One or more of the following: "alternative", "dvd", "festival", "tv", "video", "working", "original", "imdbDisplay". New values may be added in the future without warning
# attributes (array) - Additional terms to describe this alternative title, not enumerated
# isOriginalTitle (boolean) – 0: not original title; 1: original title

title_akas = pd.read_csv('title.akas.tsv',sep='\t')
title_akas.head(5)
```

```python
# title.basics.tsv.gz - Contains the following information for titles:

# tconst (string) - alphanumeric unique identifier of the title
# titleType (string) – the type/format of the title (e.g. movie, short, tvseries, tvepisode, video, etc)
# primaryTitle (string) – the more popular title / the title used by the filmmakers on promotional materials at the point of release
# originalTitle (string) - original title, in the original language
# isAdult (boolean) - 0: non-adult title; 1: adult title
# startYear (YYYY) – represents the release year of a title. In the case of TV Series, it is the series start year
# endYear (YYYY) – TV Series end year. ‘\N’ for all other title types
# runtimeMinutes – primary runtime of the title, in minutes
# genres (string array) – includes up to three genres associated with the title

title_basics = pd.read_csv('title.basics.tsv',sep='\t')
title_basics.head(5)
```

```python
# title.principals.tsv.gz – Contains the principal cast/crew for titles

# tconst (string) - alphanumeric unique identifier of the title
# ordering (integer) – a number to uniquely identify rows for a given titleId
# nconst (string) - alphanumeric unique identifier of the name/person
# category (string) - the category of job that person was in
# job (string) - the specific job title if applicable, else '\N'
# characters (string) - the name of the character played if applicable, else '\N'

title_principals = pd.read_csv('title.principals.tsv',sep='\t')
title_principals.head(5)
```

```python
'''
LIST OF GENRES
'''

genres_df = movies_main.groupby(['genres'])['tconst'].count().reset_index()
# or:
# genres_df2 = title_basics.groupby(['genres'])['tconst'].agg('count').reset_index()

genres_df_3col = genres_df['genres'].str.split(',',expand=True)
genres_list = list(set(list(genres_df_3col[0]) + list(genres_df_3col[1]) + list(genres_df_3col[2])) )
genres_list.remove(None)
genres_list.remove('\\N')
genres_list = sorted(genres_list)

genres_list
```

```python
'''
THE FIRST BIG FUNCTION: TO FILTER TITLES BY THEIR ATTRIBUTES
'''
```

```python
'''
[+] TO FIND ALL TITLES AT LEAST WITH ONE GENRE FROM THE LIST
'''
def filter_genre_by_list(df, tag_list):
    df_2 = df[df['genres'].notnull()]
    df_3 = df_2.loc[df_2['genres'].apply(lambda x: True if any(str.lower(tag) in str.lower(x) for tag in tag_list) else False)]
    return df_3
'''
[+] TO FIND ALL TITLES WITHOUT EACH GENRE FROM THE ANTI-LIST
'''
def filter_genre_by_anti_list(df, tag_list = []):
    df_2 = df[df['genres'].notnull()]
    df_3 = df_2.loc[df_2['genres'].apply(lambda x: False if any(str.lower(tag) in str.lower(x) for tag in tag_list) else True)]
    return df_3

'''
[+] TO FIND ALL TITLES CONTAINED A GIVEN LIST OF TAGS
'''
def filter_title_by_list(df, tag_list = []):
#     column_nm = 'originalTitle'
    df_2 = df[df['originalTitle'].notnull()]
    df_3 = df_2.loc[df_2['originalTitle'].apply(lambda x: True if all(str.lower(tag) in str.lower(x) for tag in tag_list) else False)]
    return df_3

'''
[+] TO FIND ALL TITLES WITH RATING MORE THAN GIVEN LIMIT
'''
def filter_rating(df, low_limit=0, min_votes=0):
    df_2 = df[df['averageRating'].apply(lambda x: True if float(x) >= low_limit else False)]
    df_3 = df_2[df_2['numVotes'].apply(lambda x: True if int(x) >= min_votes else False)]
    return df_3

'''
[+] TO FIND ALL TITLES MADE IN GIVEN YEAR RANGE
'''
def filter_year(df, left_year=1800, right_year=2200):
    df_2 = df[df['startYear'] != '\\N']
    df_3 = df_2[df_2['startYear'].apply(lambda x: True if (int(x) >= left_year and int(x) <= right_year) else False)]
    return df_3

'''
[+] TO FIND ALL TITLES WITH DURATION IN GIVEN RANGE
'''
def filter_runtime(df, left_runtime=0, right_runtime=1000):
    df_2 = df[df['runtimeMinutes'] != '\\N']
    df_3 = df_2[df_2['runtimeMinutes'].apply(lambda x: True if (int(x) >= left_runtime and int(x) <= right_runtime) else False)]
    return df_3


def filter_title_by_attributes(df, attr_dict):
    if 'genres_list' in attr_dict.keys() and 'exclude_genres_list' in attr_dict.keys() and len([g for g in attr_dict['genres_list'] if g in attr_dict['exclude_genres_list']]) > 0:
        raise Exception("Error: genres_list and exclude_genres_list crosses!")
    else:
        if 'genres_list' in attr_dict.keys():
            df = filter_genre_by_list(df, attr_dict['genres_list'])

        if 'exclude_genres_list' in attr_dict.keys():
            df = filter_genre_by_anti_list(df, attr_dict['exclude_genres_list'])

#         if attr_dict['tags_list'] is not None:
        if 'tags_list' in attr_dict.keys():
            df = filter_title_by_list(df, attr_dict['tags_list'])
            
        if 'min_rating' in attr_dict.keys() and 'min_votes' in attr_dict.keys():
            df = filter_rating(df, attr_dict['min_rating'], attr_dict['min_votes'])
        elif 'min_rating' in attr_dict.keys():
            df = filter_rating(df, attr_dict['min_rating'], 0)
        elif 'min_votes' in attr_dict.keys():
            df = filter_rating(df, 0, attr_dict['min_votes'])

        if 'min_year' in attr_dict.keys() and 'max_year' in attr_dict.keys():
            df = filter_year(df, attr_dict['min_year'], attr_dict['max_year'])
        elif 'min_year' in attr_dict.keys():
            df = filter_year(df, attr_dict['min_year'], 2200)
        elif 'max_year' in attr_dict.keys():
            df = filter_year(df, 1800, attr_dict['max_year'])

        if 'min_runtime' in attr_dict.keys() and 'max_runtime' in attr_dict.keys():
            df = filter_runtime(df, attr_dict['min_runtime'], attr_dict['max_runtime'])
        elif 'min_runtime' in attr_dict.keys():
            df = filter_runtime(df, attr_dict['min_runtime'], 1000)
        elif 'max_runtime' in attr_dict.keys():
            df = filter_runtime(df, 0, attr_dict['max_runtime'])

        if df.shape[0] > 0 and 'order_of_sorting' in attr_dict.keys():
            df = df.sort_values([s[0] for s in attr_dict['order_of_sorting']],
                  ascending = [s[1] for s in attr_dict['order_of_sorting']])
    return df
```

```python
'''
THE SECOND BIG FUNCTION: TO FILTER TITLES BY PERSONS
'''
```

```python
category_list = list(title_principals.groupby(['category'])['tconst'].count().reset_index()['category'])
category_list
```

```python
'''
[+] TO RETURN ALL target_col FOR col CONTAINED ALL/ANY VALUES FROM options (NEW -- COMMON VERSION)
'''

def filter_column_by_list(df, target_col, col, options, quantifier):
    df = df[[target_col, col]][df[col].notnull()]
    if quantifier == 'all':
        for option in options:
            df_temp = df.loc[df[col].apply(lambda x: True if str.lower(option) == str.lower(x) else False)]
            df = df.merge(df_temp[[target_col]], left_on=target_col, right_on=target_col, how='inner')
    elif quantifier == 'any':
        df_list = []
        for option in options:
            df_temp = df.loc[df[col].apply(lambda x: True if str.lower(option) == str.lower(x) else False)]
            df_list.append(df_temp[[target_col]])
        df_temp = pd.concat(df_list)
        df = df.merge(df_temp[[target_col]], left_on=target_col, right_on=target_col, how='inner')
    df = df.drop_duplicates()
    return df[[target_col]]

'''
[+] TO FIND ALL TITLES CONTAINED A GIVEN LIST OF SEVERAL CATEGORIES
'''
def filter_titles_by_persons(df, search_dict):
    if len(search_dict) == 0:
        return 'No conditions :('
    else:
        category_list = list(title_principals.groupby(['category'])['tconst'].count().reset_index()['category'])
        for category_nm in category_list:
            if category_nm in search_dict.keys():
                pers_df = title_principals[['tconst', 'nconst', 'category']][title_principals['category'] == category_nm]
                pers_df_temp = name_basics[['nconst','primaryName','birthYear','deathYear']].merge(pers_df, left_on='nconst', right_on='nconst', how='inner')
                df = df.merge(filter_column_by_list(pers_df_temp, 'tconst', 'primaryName', search_dict[category_nm][0], search_dict[category_nm][1])[['tconst']], left_on='tconst', right_on='tconst', how='inner')
        df = df.drop_duplicates()
        
        category_list = ['director', 'composer', 'producer', 'writer']
        pers_df = title_principals[['tconst', 'nconst', 'category']][title_principals['category'].isin(category_list)]
        pers_df_temp = name_basics[['nconst','primaryName']].merge(pers_df, left_on='nconst', right_on='nconst', how='inner')
        tail_df = df[['tconst']].merge(pers_df_temp[['tconst','primaryName','category']], left_on='tconst', right_on='tconst', how='inner')
        tail_df = tail_df.drop_duplicates()
        tail_df = pd.melt(tail_df, 
             id_vars=['tconst', 'primaryName'],
             value_vars=['category'], 
             value_name='category2').groupby(['tconst','category2'])['primaryName'].agg(', '.join).reset_index()
        tail_df = tail_df.drop_duplicates()
        tail_df = tail_df.pivot(index="tconst", columns="category2", values="primaryName").reset_index()
        df2 = df.merge(tail_df, left_on='tconst', right_on='tconst', how='left')
        
        return df2

        
```

```python
'''
[+] COMMON FUNCTION
'''

# need the following tables:
# title_basics
# title_ratings
# title_principals
# name_basics

def filter_imdb(filter_dict):
    if 'titleType' in filter_dict.keys():
        df_movies = title_basics[title_basics['titleType'].isin(filter_dict['titleType'])].merge(title_ratings, left_on='tconst', right_on='tconst')
    else:
        df_movies = title_basics.merge(title_ratings, left_on='tconst', right_on='tconst')
        
    df_movies = filter_title_by_attributes(df_movies, filter_dict)
    df_movies = filter_titles_by_persons(df_movies, filter_dict)
    df_movies.drop(['isAdult'], axis=1)
    
    return df_movies
```
